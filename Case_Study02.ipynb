# 3. Who scores the most on average for math, reading and writing based on - A) Gender, B)Test preparation course

## 3.(a) Based on Gender

```python
dataset.head()
math_scoreAvg=dataset.groupby('gender')['math score'].mean()
reading_scoreAvg=dataset.groupby('gender')['reading score'].mean()
writing_scoreAvg=dataset.groupby('gender')['writing score'].mean()

Avg_Scores=pd.DataFrame(math_scoreAvg)
reading_Avg_Scores=pd.DataFrame(reading_scoreAvg)
writing_Avg_Scores=pd.DataFrame(writing_scoreAvg)

extracted_col1 = reading_Avg_Scores["reading score"]
extracted_col2 = writing_Avg_Scores["writing score"]
Avg_Scores=Avg_Scores.join(extracted_col1)
Avg_Scores=Avg_Scores.join(extracted_col2)
Avg_Scores
```

### <U>Observation:</U>
Female students topped  in reading and writing tests whereas boys excelled in maths considering their average test scores

## 3.(b) Based on - Test preparation course


```python
dataset.head()
math_scoreAvg1=dataset.groupby('test preparation course')['math score'].mean()
reading_scoreAvg1=dataset.groupby('test preparation course')['reading score'].mean()
writing_scoreAvg1=dataset.groupby('test preparation course')['writing score'].mean()

Avg_Scores1=pd.DataFrame(math_scoreAvg1)
reading_Avg_Scores1=pd.DataFrame(reading_scoreAvg1)
writing_Avg_Scores1=pd.DataFrame(writing_scoreAvg1)

extracted_col2 = reading_Avg_Scores1["reading score"]
extracted_col3 = writing_Avg_Scores1["writing score"]
Avg_Scores1=Avg_Scores1.join(extracted_col2)
Avg_Scores1=Avg_Scores1.join(extracted_col3)
Avg_Scores1
```

### <u>Observation</u>:
It's undoubtly seen from the results that, those who have completed the test preparation course excelled in all the exams with better average scores than those who did'nt

# 4. What do you think about the scoring variation for math, reading and writing based on A)Gender, B)Test preparation course



## 4.(a) Based on Gender

### (i)Statistical features and Histograms of math sore

```python
#Math Score comparison - Females and Males
males=dataset[dataset["gender"]=='male']
females=dataset[dataset["gender"]=='female']
plt.figure()
males[['math score']].plot.hist(rwidth=0.9)
plt.title('Math test scores- males')
plt.xlabel('Marks scored')

females[['math score']].plot.hist(rwidth=0.9)
plt.title('Math test scores- Females')
plt.xlabel('Marks scored')

m_details=males['math score'].describe()
f_details=females['math score'].describe()
m_details=pd.DataFrame(m_details)
m_details=m_details.rename(columns={"math score":"Math score -Male students"})
f_details=pd.DataFrame(f_details)
f_details=f_details.rename(columns={"math score":"Math score -Female students"})

extracted_col6 = f_details["Math score -Female students"]
m_details=m_details.join(extracted_col6)

m_details


```

### <u>Observations</u>
Male students have better average scores in math test than females and the range of score variation is quite stable among boys(evident from the standard deviation values). So the score variation among male students is less compared to female students. Apart from the standard deviation values, the range also illustrates this.

### (ii)Statistical features and Histograms of reading score 

```python
#reading score comparison - Females and Males

plt.figure()
males[['reading score']].plot.hist(rwidth=0.9)
plt.title('reading test scores- males')
plt.xlabel('Marks scored')

females[['reading score']].plot.hist(rwidth=0.9)
plt.title('reading score scores- Females')
plt.xlabel('Marks scored')

m_details1=males['reading score'].describe()
f_details1=females['reading score'].describe()
m_details1=pd.DataFrame(m_details1)
m_details1=m_details1.rename(columns={"reading score":"reading score -Male students"})
f_details1=pd.DataFrame(f_details1)
f_details1=f_details1.rename(columns={"reading score":"reading score -Female students"})

extracted_col7 = f_details1["reading score -Female students"]
m_details1=m_details1.join(extracted_col7)

m_details1
```

### <U>Observations</U>
Female students performed well in the reading test and the score variation among females and males are almost the same with males showing a bit small variation in scores(evident from standard deviation values). The range of test score is also showing similar trend.

### (iii)Statistical features and Histograms of writing score

```python
#writing score comparison - Females and Males

plt.figure()
males[['writing score']].plot.hist(rwidth=0.9)
plt.title('writing test scores- males')
plt.xlabel('Marks scored')

females[['writing score']].plot.hist(rwidth=0.9)
plt.title('writing test score - Females')
plt.xlabel('Marks scored')

m_details2=males['writing score'].describe()
f_details2=females['writing score'].describe()
m_details2=pd.DataFrame(m_details2)
m_details2=m_details2.rename(columns={"writing score":"writing score -Male students"})
f_details2=pd.DataFrame(f_details2)
f_details2=f_details2.rename(columns={"writing score":"writing score -Female students"})

extracted_col8 = f_details2["writing score -Female students"]
m_details2=m_details2.join(extracted_col8)

m_details2
```

### <U>Observations</U>
Females performed well in writing test, with far better average score than male counterparts. But both of them have almost same level of consistency(though male students showing slightly better consistency with a small fraction, see the standard deviation values)

### (ii)Box plot

```python
males.boxplot()
plt.title("Scores- Male students")
plt.style.use('seaborn')
```

```python
females.boxplot()
plt.title("Scores- Female students")
plt.style.use('seaborn')
```

### <u>Observatios:</u>
there is a significant number of outliers in case of female students(as seen from the box plots), so better care and attention should be given to female students to improve their performance.    

## 4.(b) Based on Test preparation course

### (i)Statistical features and Histograms of math sore

```python

test_completed=dataset[dataset["test preparation course"]=='completed']
test_none=dataset[dataset["test preparation course"]=='none']
plt.figure()
test_completed[['math score']].plot.hist(rwidth=0.9)
plt.title('Math test scores- those who took test prep. course')
plt.xlabel('Marks scored')

test_none[['math score']].plot.hist(rwidth=0.9)
plt.title('Math test scores- those who did not took test prep course')
plt.xlabel('Marks scored')

completed_details=test_completed['math score'].describe()
none_details=test_none['math score'].describe()
completed_details=pd.DataFrame(completed_details)
completed_details=completed_details.rename(columns={"math score":"Math score -those who took prep course"})
none_details=pd.DataFrame(none_details)
none_details=none_details.rename(columns={"math score":"Math score -those who did not took prep course"})

extracted_col10 = none_details["Math score -those who did not took prep course"]
completed_details=completed_details.join(extracted_col10)

completed_details
```

### <u>Observations</u>
There is significant advantage for those who completed the test prep. course. It can be easily identified from the average scores and the high consistency of those who completed test prep course(by standard deviation data). 

### (ii)Statistical features and Histograms of reading score
