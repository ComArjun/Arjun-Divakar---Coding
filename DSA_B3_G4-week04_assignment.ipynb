## 1)Compare churn count with respect to gender. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataset=pd.read_excel('Telco-Customer-Churn.xlsx')
churn_yes = dataset[dataset['Churn']=='Yes']
fig, ax = plt.subplots(figsize=(6, 6))
sns.countplot(churn_yes['gender'])
plt.grid()
plt.show()
```

## 3)Compare 'tenure' with 'Total Charges' 

```python
data=pd.read_excel("Telco-Customer-Churn.xlsx")
data["TotalCharges"]=pd.to_numeric(data["TotalCharges"],downcast="float",errors='coerce')
data["TotalCharges"].fillna(data["TotalCharges"].mean(),inplace=True)
plt.figure(figsize=(18,8))
plt.bar(data['tenure'],data['TotalCharges'],color="#20B2AA")
plt.xlabel("Tenure")
plt.ylabel("Total Charges")
plt.figure(figsize=(18,18))
sns.boxplot(x='tenure',y='TotalCharges',data=data )
```

## 4)Find out which contract preffered by the senior citizen. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dataset=pd.read_excel('Telco-Customer-Churn.xlsx')
#dataset.info()
Senior_citizens = dataset[dataset['SeniorCitizen']==1]
Senior_citizens.head()
fig, ax = plt.subplots(figsize=(6, 6))
sns.countplot(Senior_citizens['Contract'])
plt.grid()
plt.show()
```

### The most preferred contract among senior citizens is Month-to Month contract followed by one& two year contracts
