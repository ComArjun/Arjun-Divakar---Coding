# 5].Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
#Visualization-1- swarmplot- Clasification Vs SL
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.swarmplot(x='Classification',y='SL',data=dataset)
```

```python
#Visualization-2 - Distribution & Kernal density of SW & PL
plt.figure()
dataset[['SW','PL']].plot.kde()
dataset[['SW','PL']].plot.hist(rwidth=0.6)
plt.title('Distribution of SW & PL')
plt.tight_layout()
```

```python
#Visualization-3- Pairplots
sns.pairplot(dataset)
```

```python
#Visualization-4 - SL Vs PW
plt.figure()
plt.scatter(dataset['SL'],dataset['PW'],s=20,marker='*')
plt.title('SL Vs PW')
plt.xlabel('SL')
plt.ylabel('PW')
```
