# 4. Carry out correlation analysis on the dataset and find the correlation factor between

### a. ph and Chloramines

### b. ph and Conductivity

### What do we infer from this correlation value?

```python
corr_matrix=dataset.corr()
plt.subplots(figsize=(15,6))
sns.heatmap(corr_matrix,vmin=-0.1,vmax=0.15,annot=True,Linewidths=0.2,cmap='YlGnBu')
```

### a. ph and Chloramines

```python
x = dataset[['ph', 'Chloramines']]
x.corr()
```

####  <u>Result</u>: The correlation factor between ph and chloramines is -(0.025)(from the correlation matrix), which is a low negative correlation and from the graph we can find that there is no linear relationship between them

```python
plt.figure(figsize=(6,4))
plt.scatter(dataset['ph'],dataset['Chloramines'],s=10,c='red')
plt.title('ph Vs Chloramines',fontsize=10)
plt.xlabel('ph',fontsize=10)
plt.ylabel('Chloramines',fontsize=10)
```

### b. ph and Conductivity

```python
x = dataset[['ph', 'Conductivity']]
x.corr()
```

#### <u>Result</u>: The correlation factor between ph and Conductivity is 0.014(from the correlation matrix) which is a low positive correlation and from the graph we can find that there is no linear relationship between them 

```python
plt.figure(figsize=(6,4))
plt.scatter(dataset['ph'],dataset['Conductivity'],s=10,c='green')
plt.title('ph Vs Conductivity',fontsize=10)
plt.xlabel('ph',fontsize=10)
plt.ylabel('Conductivity',fontsize=10)
```

### <u>Inference:</u> ph value has no strong correlation with any of the factors under study; but exhibits positive correlation with hardness(Highest among features under study),Sulfate	Conductivity,Organic_carbon,Trihalomethanes.
